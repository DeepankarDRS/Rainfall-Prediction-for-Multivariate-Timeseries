```python

import matplotlib.pyplot as plt
import seaborn as sns

# Line plot of annual rainfall over the years
plt.figure(figsize=(10, 6))
plt.plot(df1['YEAR'], df1['ANNUAL'])
plt.xlabel('Year')
plt.ylabel('Annual Rainfall')
plt.title('Annual Rainfall Trend')
plt.grid(True)
plt.show()

# Box plot of rainfall distribution for each month
plt.figure(figsize=(12, 6))
sns.boxplot(data=df1[['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC']])
plt.xlabel('Month')
plt.ylabel('Rainfall')
plt.title('Rainfall Distribution by Month')
plt.xticks(rotation=45)
plt.show()



# Correlation matrix heatmap
plt.figure(figsize=(12, 10))
correlation_matrix = df1.corr()  # Calculate the correlation matrix
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Rainfall Data')
plt.show()
```

```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objs as go
from plotly.subplots import make_subplots

# Assuming df2 is your DataFrame and it has the specified columns
# Replace 'your_dataframe' with the actual name of your DataFrame if different
columns_to_plot = ['YEAR', 'JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC', 'JF', 'MAM', 'JJAS', 'OND', 'ANN']


# Function to create the plots
def plot_rainfall_data(df, columns_to_plot):
    # Line plot of annual rainfall over the years
    plt.figure(figsize=(10, 6))
    plt.plot(df['YEAR'], df['ANN']) #plot 'ANN' column instead of 'ANNUAL'
    plt.xlabel('Year')
    plt.ylabel('Annual Rainfall')
    plt.title('Annual Rainfall Trend')
    plt.grid(True)
    plt.show()

    # Box plot of rainfall distribution for each month
    plt.figure(figsize=(12, 6))
    sns.boxplot(data=df[['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC']])
    plt.xlabel('Month')
    plt.ylabel('Rainfall')
    plt.title('Rainfall Distribution by Month')
    plt.xticks(rotation=45)
    plt.show()

    # Correlation matrix heatmap
    plt.figure(figsize=(12, 10))
    correlation_matrix = df[columns_to_plot].corr()  # Calculate the correlation matrix, only with specified columns
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
    plt.title('Correlation Matrix of Rainfall Data')
    plt.show()

#Call the function with your dataframe and specified columns

plot_rainfall_data(df2, columns_to_plot)
```

```python
import pandas as pd
from sklearn.preprocessing import RobustScaler, MinMaxScaler

year_col = df2["YEAR"]

# Define column groups
monthly_cols = ['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC']
seasonal_cols = ['JF', 'MAM', 'JJAS', 'OND', 'ANN']

# -----------------------
# 1. Min-Max Scaling for all columns (monthly + seasonal)
# -----------------------
minmax_scaler = MinMaxScaler()

# Apply Min-Max Scaling to both monthly and seasonal columns
monthly_minmax = minmax_scaler.fit_transform(df2[monthly_cols])
seasonal_minmax = minmax_scaler.fit_transform(df2[seasonal_cols])

# Convert to DataFrame
monthly_minmax_df = pd.DataFrame(monthly_minmax, columns=monthly_cols)
seasonal_minmax_df = pd.DataFrame(seasonal_minmax, columns=seasonal_cols)

# Combine and add YEAR column
minmax_final_df = pd.concat([year_col, monthly_minmax_df, seasonal_minmax_df], axis=1)

# Save Min-Max scaled data
minmax_final_df.to_csv("minmax_scaled_rainfall.csv", index=False)

# -----------------------
# 2. Robust Scaling for monthly columns (no Min-Max Scaling for seasonal)
# -----------------------

# Apply Robust Scaling to monthly columns
robust_scaler = RobustScaler()
monthly_robust = robust_scaler.fit_transform(df2[monthly_cols])

# Convert to DataFrame
monthly_robust_df = pd.DataFrame(monthly_robust, columns=monthly_cols)

# Combine and add YEAR column (seasonal columns are unchanged)
robust_final_df = pd.concat([year_col, monthly_robust_df, df2[seasonal_cols]], axis=1)

# Save Robust scaled data
robust_final_df.to_csv("robust_scaled_rainfall.csv", index=False)

# -----------------------
# Confirmation
# -----------------------
print("✅ Min-Max scaled dataset saved as 'minmax_scaled_rainfall.csv'")
print("✅ Robust scaled dataset saved as 'robust_scaled_rainfall.csv'")
```
