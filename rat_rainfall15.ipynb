```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping

# ====================== 1. Data Preparation ======================
def prepare_data(mdf, target_col='ANN_x', window_size=5, step_size=7, test_size=0.2):
    """
    Prepares time-series data for LSTM training.
    Returns:
    X_train, X_test, y_train, y_test, scaler, X (full sequences)
    """
    # Normalize each feature between 0-1
    scaler = MinMaxScaler()
    scaled_data = scaler.fit_transform(mdf)

    # Create sequences (5 years ‚Üí predict next year)
    X, y = [], []
    for i in range(0, len(scaled_data) - window_size - 1, step_size):
        X.append(scaled_data[i:i+window_size])
        y.append(scaled_data[i+window_size, mdf.columns.get_loc(target_col)])

    # Convert to numpy arrays
    X, y = np.array(X), np.array(y)

    # Sequential train-test split (no shuffling)
    train_size = int(len(X) * (1 - test_size))
    X_train, X_test = X[:train_size], X[train_size:]
    y_train, y_test = y[:train_size], y[train_size:]

    return X_train, X_test, y_train, y_test, scaler, X  # Now returning X
# ====================== 2. LSTM Model ======================
def build_lstm_model(input_shape):
    """Builds an LSTM model with dropout regularization"""
    model = Sequential([
        LSTM(64, return_sequences=True, input_shape=input_shape),
        Dropout(0.2),
        LSTM(32),
        Dropout(0.2),
        Dense(1)
    ])

    model.compile(
        optimizer=Adam(learning_rate=0.001),
        loss='mse',  # MSE works better for regression
        metrics=['mae']
    )
    return model

# ====================== 3. Training ======================
def train_model(X_train, y_train, X_test, y_test, epochs=500, batch_size=8):
    """Trains the LSTM with early stopping"""
    early_stop = EarlyStopping(monitor='val_loss', patience=10)

    model = build_lstm_model((X_train.shape[1], X_train.shape[2]))

    history = model.fit(
        X_train, y_train,
        epochs=epochs,
        batch_size=batch_size,
        validation_data=(X_test, y_test),
        callbacks=[early_stop],
        verbose=1
    )

    return model, history

# ====================== 4. Evaluation ======================
def evaluate_model(model, X_test, y_test, scaler, target_col_index):
    """Evaluates model and plots results"""
    # Predictions
    y_pred = model.predict(X_test)

    # Inverse transform predictions
    dummy = np.zeros((len(y_pred), X_test.shape[2]))
    dummy[:, target_col_index] = y_pred.flatten()
    y_pred = scaler.inverse_transform(dummy)[:, target_col_index]

    # Inverse transform actual values
    dummy[:, target_col_index] = y_test.flatten()
    y_actual = scaler.inverse_transform(dummy)[:, target_col_index]

    # Metrics
    mae = mean_absolute_error(y_actual, y_pred)
    rmse = np.sqrt(mean_squared_error(y_actual, y_pred))
    r2 = r2_score(y_actual, y_pred)

    # Plot
    plt.figure(figsize=(12, 6))
    plt.plot(y_actual, label='Actual', marker='o')
    plt.plot(y_pred, label='Predicted', marker='x')
    plt.title('Actual vs Predicted Annual Rainfall')
    plt.xlabel('Time Steps')
    plt.ylabel('Rainfall (mm)')
    plt.legend()
    plt.grid(True)
    plt.show()

    print(f"MAE: {mae:.2f}, RMSE: {rmse:.2f}, R¬≤: {r2:.2f}")
    return mae, rmse, r2, y_actual, y_pred

# ====================== 5. Future Predictions ======================
def predict_future(model, last_sequence, scaler, target_col_index, steps=5):
    """Predicts future rainfall beyond the dataset"""
    predictions = []
    current_sequence = last_sequence.copy()

    for _ in range(steps):
        pred = model.predict(current_sequence[np.newaxis, :, :])[0, 0]

        # Update sequence (shift left and append prediction)
        new_sequence = np.roll(current_sequence, -1, axis=0)
        new_sequence[-1, target_col_index] = pred
        current_sequence = new_sequence

        # Inverse transform
        dummy = np.zeros((1, current_sequence.shape[1]))
        dummy[0, target_col_index] = pred
        predictions.append(scaler.inverse_transform(dummy)[0, target_col_index])

    return predictions

# ====================== MAIN EXECUTION ======================
if __name__ == "__main__":
    # Load your data (replace with your DataFrame)
    # mdf = pd.read_csv("your_data.csv")

    # Example usage:
    target_col = 'ANN_x'
    target_col_index = mdf.columns.get_loc(target_col)

    # 1. Prepare data - now returning X for future predictions
    X_train, X_test, y_train, y_test, scaler, X = prepare_data(
        mdf, target_col=target_col, window_size=5, step_size=7, test_size=0.2
    )

    # 2. Train model
    model, history = train_model(X_train, y_train, X_test, y_test)

    # 3. Evaluate
    mae, rmse, r2, y_actual, y_pred = evaluate_model(
        model, X_test, y_test, scaler, target_col_index
    )

    # 4. Future predictions - now using the correct X
    last_sequence = X[-1]  # Most recent window
    future_rainfall = predict_future(
        model, last_sequence, scaler, target_col_index, steps=5
    )
    print(f"Next 5 years rainfall predictions: {future_rainfall}")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, Input
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping

# ====================== 1. Data Preparation ======================
def prepare_data(mdf, target_col='ANN_x', window_size=5, step_size=1, test_size=0.2):
    """Prepares time-series data for LSTM training with improved windowing"""
    # Normalize each feature between 0-1
    scaler = MinMaxScaler()
    scaled_data = scaler.fit_transform(mdf)

    # Create sequences with sliding window
    X, y = [], []
    for i in range(0, len(scaled_data) - window_size, step_size):
        X.append(scaled_data[i:i+window_size])
        y.append(scaled_data[i+window_size, mdf.columns.get_loc(target_col)])

    X, y = np.array(X), np.array(y)

    # Sequential train-test split
    train_size = int(len(X) * (1 - test_size))
    X_train, X_test = X[:train_size], X[train_size:]
    y_train, y_test = y[:train_size], y[train_size:]

    return X_train, X_test, y_train, y_test, scaler, X

# ====================== 2. Improved LSTM Model ======================
def build_lstm_model(input_shape):
    """Enhanced model architecture with better regularization"""
    model = Sequential([
        Input(shape=input_shape),  # Explicit input layer
        LSTM(128, return_sequences=True),
        Dropout(0.3),
        LSTM(64),
        Dropout(0.3),
        Dense(32, activation='relu'),
        Dense(1)
    ])

    model.compile(
        optimizer=Adam(learning_rate=0.0005),
        loss='huber_loss',  # Robust to outliers
        metrics=['mae']
    )
    return model

# ====================== 3. Training with Enhanced Settings ======================
def train_model(X_train, y_train, X_test, y_test):
    """Training process with better early stopping"""
    early_stop = EarlyStopping(
        monitor='val_loss',
        patience=20,
        restore_best_weights=True,
        verbose=1
    )

    model = build_lstm_model((X_train.shape[1], X_train.shape[2]))

    history = model.fit(
        X_train, y_train,
        epochs=200,
        batch_size=16,
        validation_data=(X_test, y_test),
        callbacks=[early_stop],
        verbose=1
    )

    # Plot training history
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title('Training History: Loss')
    plt.ylabel('Loss')
    plt.xlabel('Epoch')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(history.history['mae'], label='Train MAE')
    plt.plot(history.history['val_mae'], label='Validation MAE')
    plt.title('Training History: MAE')
    plt.ylabel('MAE')
    plt.xlabel('Epoch')
    plt.legend()

    plt.tight_layout()
    plt.show()

    return model, history

# ====================== 4. Enhanced Evaluation ======================
def evaluate_model(model, X_test, y_test, scaler, target_col_index):
    """Improved evaluation with better inverse scaling"""
    y_pred = model.predict(X_test, verbose=0)

    # Inverse transform predictions
    dummy = np.zeros((len(y_pred), X_test.shape[2]))
    dummy[:, target_col_index] = y_pred.flatten()
    y_pred = scaler.inverse_transform(dummy)[:, target_col_index]

    # Inverse transform actual values
    dummy[:, target_col_index] = y_test.flatten()
    y_actual = scaler.inverse_transform(dummy)[:, target_col_index]

    # Calculate metrics
    mae = mean_absolute_error(y_actual, y_pred)
    rmse = np.sqrt(mean_squared_error(y_actual, y_pred))
    r2 = r2_score(y_actual, y_pred)

    # Plot results
    plt.figure(figsize=(12, 6))
    plt.plot(y_actual, label='Actual', marker='o', color='blue')
    plt.plot(y_pred, label='Predicted', marker='x', color='red')
    plt.title('Actual vs Predicted Rainfall')
    plt.xlabel('Time Steps')
    plt.ylabel('Rainfall (mm)')
    plt.legend()
    plt.grid(True)
    plt.show()

    print(f"Performance Metrics:")
    print(f"MAE: {mae:.2f}")
    print(f"RMSE: {rmse:.2f}")
    print(f"R¬≤: {r2:.2f}")

    return mae, rmse, r2, y_actual, y_pred

# ====================== 5. Robust Future Predictions ======================
def predict_future(model, last_sequence, scaler, target_col_index, steps=5):
    """Improved future prediction with sequence updating"""
    predictions = []
    current_sequence = last_sequence.copy()

    for _ in range(steps):
        # Predict next step
        pred = model.predict(current_sequence[np.newaxis, :, :], verbose=0)[0, 0]

        # Update sequence (shift and replace target)
        new_sequence = np.roll(current_sequence, -1, axis=0)
        new_sequence[-1, target_col_index] = pred
        current_sequence = new_sequence

        # Inverse transform
        dummy = np.zeros((1, current_sequence.shape[1]))
        dummy[0, target_col_index] = pred
        predictions.append(scaler.inverse_transform(dummy)[0, target_col_index])

    return predictions

# ====================== MAIN EXECUTION ======================
if __name__ == "__main__":
    # Load your data (replace with your actual data loading)
    # mdf = pd.read_csv("your_data.csv")

    # Configuration
    target_col = 'ANN_x'
    target_col_index = mdf.columns.get_loc(target_col)

    # 1. Prepare data
    print("Preparing data...")
    X_train, X_test, y_train, y_test, scaler, X = prepare_data(
        mdf, target_col=target_col, window_size=5, step_size=1
    )

    # 2. Train model
    print("\nTraining model...")
    model, history = train_model(X_train, y_train, X_test, y_test)

    # 3. Evaluate
    print("\nEvaluating model...")
    mae, rmse, r2, y_actual, y_pred = evaluate_model(
        model, X_test, y_test, scaler, target_col_index
    )

    # 4. Future predictions
    print("\nGenerating future predictions...")
    last_sequence = X[-1]  # Most recent window
    future_rainfall = predict_future(
        model, last_sequence, scaler, target_col_index, steps=5
    )

    print("\nNext 5 years rainfall predictions:")
    for i, pred in enumerate(future_rainfall, 1):
        print(f"Year {i}: {pred:.2f} mm")
```
